```python
from sklearn.model_selection import train_test_split-0
```

```python
output_data = new_data['median_house_value']
```

```python
new_data.drop(['median_house_value'], axis=1, inplace=True)
new_data.head()
```

```python
X.head()
```

```python
x_test
```

```python
X.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, Y)
```

```python
prediction = model.predict(x_test)
```

```python
prediction
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
rmse = np.sqrt(mean_squared_error(prediction, y_test))
rmse
```

```python
plt.scatter(prediction, y_test)
plt.show()
```

```python
coefficients = model.coef_
coefficients
```

```python
list_coefficients = pd.Series(model.coef_, X.columns)
list_coefficients
```
